```python
import requests
from bs4 import BeautifulSoup
result = ''
# Step 3: Provide the URL
url = "https://www.bbc.com/news/articles/ce442577v35o"

# Step 4: Make a request to fetch the HTML content
response = requests.get(url)

# Step 5: Check if the request was successful (status code 200 means OK)
if response.status_code == 200:
    html_content = response.text  # or response.content for bytes
    # Step 6: Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(html_content, "html.parser")
    # print(soup)
    # Step 7: Extract information (Example: extract all paragraphs)
    paragraphs = soup.find_all('body')
    for para in paragraphs:
        result +=para.get_text()  # Prints each paragraph's text
    print(result)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
from transformers import pipeline

# Load the zero-shot classification model
classifier = pipeline("zero-shot-classification")

```

```python
# Define your categories
categories = ["terrorism / riot", "positive / uplifting", "natural disaster", "politics", "others"]

# Classify new titles
titles = [
    "Massive earthquake shakes the city",
    "New government policy reform announced",
    "Breaking news: terrorist attack in the capital",
    "Community outreach program launched",
    "Gay porn is harmful for health",
    "Drinking is bad for you",
    "Riot in Bangaldesh"
]

# predictions = classifier(titles[0], candidate_labels=categories)

# predictions['labels'][0]
# for title, prediction in zip(titles, predictions):
#     print(f"Title: {title}\nPredicted Category: {prediction['labels'][0]} (Confidence: {prediction['scores'][0]:.2f})\n")
```

```python
predictions = classifier(titles, candidate_labels=categories)

for title, prediction in zip(titles, predictions):
    print(f"Title: {title}\nPredicted Category: {prediction['labels'][0]} (Confidence: {prediction['scores'][0]:.2f})\n")
```
